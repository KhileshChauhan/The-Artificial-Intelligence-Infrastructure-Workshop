```python
#read the text file Gettysburg Address
inputText = sc.textFile("/FileStore/tables/GettysburgAddress.txt")
```

```python
print("Show the count of lines in input text")
inputText.count()
```

```python
print("Print the first line of the text")
inputText.first()
```

```python
print("Print the first 3 lines of the text")
inputText.take(3)
```

```python
#filter the lines with the word great
linesWithGreatWord= inputText.filter(lambda line: "great" in line)

print("Print the count of lines where the word Great appears")
linesWithGreatWord.count()
```

```python
print("Print the lines where the word Great appears")
linesWithGreatWord.collect()
```

```python
#split the paragraph into various words
wordlist = inputText.flatMap(lambda line: line.split())
#create a word pair combination
wordPairs = wordlist.map(lambda s: (s, 1))
wordPairs.collect()
```

```python
#add up the word pairs and create a word by value list
wordPairsCount = wordPairs.reduceByKey(lambda a, b: a + b)
#sort the list of the word pair key count  in alphabetical order
print("Printing the word pair key in alphabetical order")
wordPairsCount.sortByKey().collect()
```
