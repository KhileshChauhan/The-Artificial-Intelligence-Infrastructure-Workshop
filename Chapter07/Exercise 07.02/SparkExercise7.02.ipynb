```
# exercise to show transformation of map()
# temperature of California for the month of July 2019
# https://www.accuweather.com/en/us/california/20619/july-weather/2121532?year=2019 
inTempMonth1inC=sc.parallelize([27,32,32,31,31,32,28,25,28,29,32,31,30,33,30,33,35,33,34,35,36,35,25,27,28,28,27,21,33,33,21])
print("Temperature of California for the month of July 2019")
outTempMonth1inC = inTempMonth1inC.map(lambda x: x).collect() 
for x in outTempMonth1inC:
  print (x)
```

```
# converting the temperature to F
outTempMonth1inF = inTempMonth1inC.map(lambda x: ((x * 9/5)+32) ).collect() 
for x in outTempMonth1inF:
  print (x)
```

```
#finding the days with temperature greater than or equal to 30 degree Celsius
outTempHighinC = inTempMonth1inC.filter(lambda x: (x >= 30)).collect()
print("Days with temperature greater than or equal to 30 degree Celsius")
for x in outTempHighinC:
  print (x)
```

```
#finding the days with temperature less than or equal to 25 degree Celsius
print("Days with temperature less than or equal to 25 degree Celsius")
outTempLowinC = inTempMonth1inC.filter(lambda x: (x <= 25)).collect()
for x in outTempLowinC:
  print (x)
```
