```python
# read cat data
import json

cat = json.load(open('../Data/US_category_id.json', 'r'))
cat
```

```python
# transform data
# step 1
import pandas as pd
df_cat = pd.DataFrame(cat)
df_cat
```

```python
# step 2
df_cat['category'] = df_cat['items'].apply(lambda x: x['snippet']['title'])
df_cat['id'] = df_cat['items'].apply(lambda x: int(x['id']))
df_cat
```

```python
# optional step
df_cat_drop = df_cat.drop(columns=['kind', 'etag', 'items'])
df_cat_drop
```

```python
# read video data
df_vids = pd.read_csv("../Data/USvideos.csv.zip", compression='zip').query('trending_date=="17.14.11"')
# merge
df_join = df_vids.merge(df_cat_drop, left_on='category_id', right_on='id')[
    ['title', 'channel_title', 'category_id', 'category', 'views']]
df_join.head(10)
```

```python
df_join.groupby('category')[['views']].sum().sort_values('views', ascending=False).head(10)
```
